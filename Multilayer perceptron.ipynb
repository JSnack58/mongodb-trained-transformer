```python
# map vocabulary to indicies
chars = sorted(list(set(''.join(words))))
string_to_index = {string:index+1 for index, string in enumerate(chars)}
string_to_index['.'] = 0
index_to_string = {index:string for string, index in string_to_index.items()}
print(index_to_string)
```

```python
#  build dataset
# number of characters to predict the next character
context_length = 3
# X holds the context, Y holds the next char
X, Y = [], []
for name in names:
    print(name)
#   holds previous chars' indicies, essentially intialize to ['.', '.', '.', ..., '.']
    context = [0] * context_length
    for char in name + '.':
        char_index = string_to_index[char]      
        X.append(context)
        Y.append(char_index)
#       view the context
#         print(''.join(index_to_string[i] for i in context), '----->', index_to_string[char_index])
#       pop oldest char index and append newest char index
        context = context[1:] + [char_index]
    print()
X = torch.tensor(X)
Y = torch.tensor(Y)
```

```python
# Indexing by X basicallly creates a X.shape shaped tensor and replaces all X's lowest elements(in this case X[i][j]) 
# with tensors from C[X[i][j]]
# ex: X = [[0,0,0], [0,1,0]] C = [ [-2.9206, -0.8237], [-1.0218, -1.5443] ], 
# C[X]    = torch.tensor(X.shape) <--- C[X[i][j]]
#                 = [ [C[0], C[0], C[0]], [C[0], C[1], C[0]] ]
#                 = [ [ [-2.9206, -0.8237], [-2.9206, -0.8237], [-2.9206, -0.8237] ],
#                     [ [-2.9206, -0.8237], [-1.0218, -1.5443], [-2.9206, -0.8237]]
#                   ] 
Embeddings = C[X]
Embeddings
```

```python
# We want to create the function tanh(Embeddings @ weights_1 + bias_1).
weights_1 = torch.rand((6,100))
bias_1 = torch.rand(100)
```

```python
# Matrix multiplication won't work because Embeddings.shape[1] != weights_1.shape[0].
Embeddings.shape
```

```python
# We separate Embedding into 3 (32 x 2)matricies then concatenate them
# to make one (32 x 6) matrix. Now Embeddings.shape[1] != weights_1.shape[0].
# Use torch.cat() is not flexible, for it to work as below the you'd need to 
# separate Embedding by a number equal to the context_length. Meaning that there
# need to be Embeddings below to concatenate properly.
torch.cat([Embeddings[:, 0, :], Embeddings[:, 1, :], Embeddings[:, 2, :]], dim=1)
```

```python
# Torch.unbind() is more flexible as it remove a matrix's dimension and
# returns multiple tensors. Essentially what we did above but scalable.
# DRAWBACK: high memory usage
torch.cat(torch.unbind(Embeddings, dim = 1), dim = 1)
```

```python
# Best method is use view, it will handle all the above for in one call
Embeddings.view(32,6)
```

```python
tanh_layer_output = torch.tanh(Embeddings.view(Embeddings.shape[0], 6) @ weights_1 + bias_1)
```

```python
tanh_layer_output.shape
```

```python
# want tanh_layer_output @ output_weights + output_bias
# (32, 100) @ (100, 27) + (1, 27) --> (32, 27).sum(dim=0) ---> (1, 27) 
output_weights = torch.rand((100, 27))
output_bias = torch.rand(27)
```

```python
logits = tanh_layer_output @ output_weights + output_bias
logits.shape, logits
```

```python
# classification requires exp
context_counts = logits.exp()
context_counts.shape, context_counts
```

```python
context_probabilities = context_counts / context_counts.sum(dim = 1, keepdims=True)
context_probabilities.shape, context_probabilities
```

```python
# Just like C[X] we'll use torch.arange(32) & Y to index context_probabilities
loss = -context_probabilities[torch.arange(32), Y].log().mean()
loss
```

```python
# Set each parameter to requries_grad to True to be able to update the weights.
for parameter in parameters:
    parameter.requires_grad = True
```

```python
# training split, dev split, test split
# 80%. 10%, 10%
import random
def build_dataset(words):
    # number of characters to predict the next character
    context_length = 3
    # X holds the context, Y holds the next char
    X, Y = [], []
    for name in names:
        print(name)
    #   holds previous chars' indicies, essentially intialize to ['.', '.', '.', ..., '.']
        context = [0] * context_length
        for char in name + '.':
            char_index = string_to_index[char]      
            X.append(context)
            Y.append(char_index)
    #       view the context
    #         print(''.join(index_to_string[i] for i in context), '----->', index_to_string[char_index])
    #       pop oldest char index and append newest char index
            context = context[1:] + [char_index]
        print()
    X = torch.tensor(X)
    Y = torch.tensor(Y)
    return X, Y
random.seed(42)
random.shuffle(words)
n1 = int(0.8 * len(words))
n2 = int(0.9 * len(words))
train_X, train_Y = build_dataset(words[:n1])
dev_X, dev_Y = build_dataset(words[n1:n2])
test_X, test_Y = build_dataset(words[n2:])
```

```python
# Sampling model
generator = torch.Generator().manual_seed(2147483647 + 10)

for _ in range(20):
    names = []
    context = [0] * context_length
    while True:
        Embeddings = C[torch.tensor([context])] # (32, 3, 2)
        tanh_layer_output = torch.tanh(Embeddings.view(1, -1) @ weights_1 + bias_1)
        logits = tanh_layer_output @ weights_2 + bias_2
        char_pobabilities = F.softmax(logits, dim = 1)
        index = torch.multinomial(char_pobabilities, num_samples = 1, generator = generator).item()
        context = context[1:] + [index]
        names.append(index)
        if index == 0:
            break
    print(''.join(index_to_string[i] for i in names))
```
